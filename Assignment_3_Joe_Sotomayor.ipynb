```python
import requests
import pandas as pd
from pprint import pprint
url = 'https://api.tidesandcurrents.noaa.gov/api/prod/datagetter?begin_date=20200101&end_date=20200110&station=8518750&product=hourly_height&datum=MLLW&time_zone=lst&units=metric&application=DataAPI_Sample&format=json'

res = requests.get(url)

data = res.json()

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

df = pd.json_normalize(data, record_path =['data'])

print(df)
```
