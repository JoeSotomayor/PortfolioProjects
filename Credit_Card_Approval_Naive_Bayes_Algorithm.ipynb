```python
import pandas as pd

df = pd.read_csv('credit_card_dataset_BI.csv')
df.head(100)
```

```python
# Cleaning the dataset to use a classification algorithm, removing the continous variable 'Annual Income' and 'ID' that doesn't make any change in predictions. 

df.drop(['Annual Income', 'ID'], axis='columns', inplace=True)
df.head()
```

```python
# Dropping the column that I want to predict: 'Status'.

inputs = df.drop('Status', axis='columns')
target = df.Status
```

```python
# Converting to 0 and 1 my Gender and House Owner columns to help the model be more efficient.

dummies = pd.get_dummies(inputs)
dummies.head(3)
```

```python
# Appending the dummies columns. 

inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head(3)
```

```python
# Dropping unnecesary columns. 

inputs.drop(['Gender ','House Owner','Gender _female','House Owner_no'], axis='columns', inplace=True)
inputs.head(3)
```

```python
# Renaming the new colums.

inputs.rename(columns = {'Gender _male':'Gender','House Owner_yes':'House Owner'}, inplace=True)
inputs.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size = 0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
model.score(X_test, y_test)
```
